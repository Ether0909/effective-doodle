### Question 1

**1.** Where are developers located?

```python
count_vals = df.Country.value_counts()#Provide a pandas series of the counts for each Country

(count_vals[:10]/df.shape[0]).plot(kind="bar"); # The below bar chart of the proportion of the top 10 countries.
plt.title("Country");
```

Conclusion:More than 20% developers are from US, around 10% from India, 10% from UK. 

### Question 2

**2.** What type of developers there are?

```python
status_vals = df.Professional.value_counts() #Provide a pandas series of the counts for each Professional status

(status_vals/df.shape[0]).plot(kind="bar");
# Below bar chart shows the proportion of individuals in each professional category
plt.title("Type of Developers");
```

Conclusion: More than 70% devlopers are preofessional or say full time. less than 30% are students or part time.
